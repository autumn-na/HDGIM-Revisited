```python
# Experiment 6: Similarity Distribution of full-precision bound hypervector

import dna
import hdgim
import torch
import matplotlib.pyplot as plt

torch.set_printoptions(sci_mode=False)

# [begin] hyperparameters
HYPERVECTOR_DIMENSION = 1000  # denoted D in paper
DNA_SEQUENCE_LENGTH = 1000
DNA_SUBSEQUENCE_LENGTH = 20  # denoted n in paper
BIT_PRECISION = 2  # denoted B in paper
NOISE_PROBABILTY = 0.1  # denoted p in paper, probability: 0 ~ 1, simulate FeFET noise
EPOCHES = 10
FUNCTION = 'cosine'
LEARNING_RATE = 10000
THRESHOLD = -0.07
# [end] hyperparameters

# create dataset
dataset = dna.DNADataset(DNA_SEQUENCE_LENGTH, DNA_SUBSEQUENCE_LENGTH)

# create model
model = hdgim.HDGIM(HYPERVECTOR_DIMENSION, DNA_SEQUENCE_LENGTH, DNA_SUBSEQUENCE_LENGTH, BIT_PRECISION, NOISE_PROBABILTY)
model.set_dataset(dataset)
model.create_voltage_matrix()
model.create_base_hypervectors()
model.create_dna_sequence()
model.create_dna_subsequences()
model.bind()
model.quantize_min_max()
model.noise()

# train
accuracies, true_similarities, false_similarities = model.train(EPOCHES, LEARNING_RATE, THRESHOLD, FUNCTION, True, True, True)

# draw histogram which shows true similarity distribution and false similarity distribution
plt.hist(true_similarities[EPOCHES-1], bins=30, density=True, alpha=0.5, color='blue', edgecolor='black', label='true')
plt.hist(false_similarities[EPOCHES-1], bins=30, density=True, alpha=0.5, color='red', edgecolor='black', label='false')

# add explanation of color
plt.legend(prop={'size': 10})

plt.title('Similarity Distribution Graph')
plt.xlabel('Similarity')
plt.ylabel('Density')
plt.show()
```

```python
# Experiment 9: CDF Test, No Noise

import dna
import hdgim
import torch
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

torch.set_printoptions(sci_mode=False)

# [begin] hyperparameters
HYPERVECTOR_DIMENSION = 1000  # denoted D in paper
DNA_SEQUENCE_LENGTH = 1000
DNA_SUBSEQUENCE_LENGTH = 20  # denoted n in paper
BIT_PRECISION = 2  # denoted B in paper
NOISE_PROBABILTY = 0.1  # denoted p in paper, probability: 0 ~ 1, simulate FeFET noise
EPOCHES = 10
FUNCTION = 'hamming'
LEARNING_RATE = 10
THRESHOLD = -131
# [end] hyperparameters

# create dataset
dataset = dna.DNADataset(DNA_SEQUENCE_LENGTH, DNA_SUBSEQUENCE_LENGTH)

# create model
model = hdgim.HDGIM(HYPERVECTOR_DIMENSION, DNA_SEQUENCE_LENGTH, DNA_SUBSEQUENCE_LENGTH, BIT_PRECISION, 0)
model.set_dataset(dataset)
model.create_base_hypervectors()
model.create_dna_sequence()
model.create_dna_subsequences()
model.bind()
# model.noise()

sorted_encoded_hypervector, indices = torch.sort(model.encoded_hypervector)
np_sorted_encoded_hypervector = sorted_encoded_hypervector.numpy()
np_normalized_sorted_encoded_hypervector = (np_sorted_encoded_hypervector - np_sorted_encoded_hypervector.mean()) / np_sorted_encoded_hypervector.std()
binary_width = 1.0 / 2**BIT_PRECISION

rv = stats.norm(0, 1) # assume standard normal distribution
yl = rv.pdf(np_normalized_sorted_encoded_hypervector)
cdf = rv.cdf(np_normalized_sorted_encoded_hypervector)
quantized_cdf = np.floor(cdf / binary_width)
quantized_hypervector = torch.from_numpy(quantized_cdf)[torch.argsort(indices)]

# draw plot
fig, axes = plt.subplots(5, 2, figsize=(10, 15))

axes[0, 0].plot(np_sorted_encoded_hypervector)
axes[0, 0].set_title("Result Value of Sorted Encoded Hypervector")
axes[0, 0].set_xlabel('Index')
axes[0, 0].set_ylabel('Value')

axes[0, 1].hist(np_sorted_encoded_hypervector, bins=30, density=True, color='blue', edgecolor='black', label='true')
axes[0, 1].set_title("Result Distribution of Sorted Encoded Hypervector")
axes[0, 1].set_xlabel('Value')
axes[0, 1].set_ylabel('Density')

axes[1, 0].plot(np_normalized_sorted_encoded_hypervector, yl)
axes[1, 0].set_title('Result Value of PDF')
axes[1, 0].set_xlabel('Value (Normalized)')
axes[1, 0].set_ylabel('Probablity')

axes[1, 1].hist(yl, bins=30, density=True, color='blue', edgecolor='black', label='true')
axes[1, 1].set_title('Result Distribution of PDF')
axes[1, 1].set_xlabel('Probablity')
axes[1, 1].set_ylabel('Density')

axes[2, 0].plot(cdf)
axes[2, 0].set_title('Result Value of CDF')
axes[2, 0].set_xlabel('Index')
axes[2, 0].set_ylabel('Probability')

axes[2, 1].hist(cdf, bins=30, density=True, color='blue', edgecolor='black', label='true')
axes[2, 1].set_title('Result Distribution of CDF')
axes[2, 1].set_xlabel('Probablity')
axes[2, 1].set_ylabel('Density')

axes[3, 0].plot(quantized_cdf)
axes[3, 0].set_title('Result Quantized Value of CDF')
axes[3, 0].set_xlabel('Value')
axes[3, 0].set_ylabel('Probability')

axes[3, 1].hist(quantized_cdf, bins=30, density=True, color='blue', edgecolor='black', label='true')
axes[3, 1].set_title('Result Distribution of Quantized CDF')
axes[3, 1].set_xlabel('Value')
axes[3, 1].set_ylabel('Density')

axes[4, 0].plot(quantized_hypervector)
axes[4, 0].set_title('Result Value of Quantized Hypervector')
axes[4, 0].set_xlabel('Value')
axes[4, 0].set_ylabel('Probability')

axes[4, 1].hist(quantized_hypervector, bins=30, density=True, color='blue', edgecolor='black', label='true')
axes[4, 1].set_title('Result Distribution of Quantized CDF')
axes[4, 1].set_xlabel('Value')
axes[4, 1].set_ylabel('Density')

# Add a title for the entire figure
plt.suptitle('Value of Unsorted/Sorted Hypervectors')

# Adjust layout to prevent clipping of titles
plt.tight_layout(rect=[0, 0, 1, 0.96])
```
